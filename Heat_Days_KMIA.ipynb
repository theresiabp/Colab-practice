```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
# Read the CSV file
df = pd.read_csv('Heat-Stats_KMIA.csv')

# # Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month']= df['Date'].dt.month
```

```python
# Assuming your CSV has a 'Date' column in the format 'YYYY-MM-DD'
df['Date'] = pd.to_datetime(df['Date'])


# Create a new column 'Month' to extract the month
df['Month'] = df['Date'].dt.strftime('%b')

# Define the temperature threshold
temperature_threshold = 90

# Calculate the number of days above the threshold for each month and year
monthly_counts = df[df['Air_Temp'] > temperature_threshold].groupby(['Year', 'Month'])['Date'].count().unstack()

# Sort the DataFrame by month in ascending order
month_order = ['Jun', 'Jul', 'Aug', 'Sep']
monthly_counts = monthly_counts[month_order]

# Create a stacked bar chart with whole number values displayed on top
fig, ax = plt.subplots(figsize=(10, 6))

colors = plt.cm.get_cmap('viridis', len(month_order))
monthly_counts.plot(kind='bar', stacked=True, ax=ax, color=[colors(i) for i in range(len(month_order))])

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Number of Days')
plt.title('Number of Days above {}°F at KMIA during Summer (2019-2023)'.format(temperature_threshold))


# Show whole number values on top of each segment (skip if value is 0)
for patch in ax.patches:
    width, height = patch.get_width(), patch.get_height()
    x, y = patch.get_xy()
    value = int(height)
    if value > 0:
        ax.annotate(value, (x + width/2, y + height/2), ha='center', va='center', fontsize=10, color='white')

# Show the legend
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.xticks(rotation= 0)
plt.show()
```

```python
df4 = pd.read_csv('Heat-Stats_KMIA.csv')

# Assuming your CSV has a 'Date' column in the format 'YYYY-MM-DD'
df4['Date'] = pd.to_datetime(df4['Date'])
df4['Year'] = df4['Date'].dt.year

# Filter data for summer months (June-September)
df4 = df4[(df4['Date'].dt.month >= 6) & (df4['Date'].dt.month <= 9)]

# Define the temperature thresholds
temperature_thresholds = [95, 100, 105]

# Create a new column 'Month' to extract the month
df4['Month'] = df4['Date'].dt.strftime('%b')

# Create a stacked bar chart for each temperature threshold
fig, ax = plt.subplots(figsize=(10, 6))

colors = plt.cm.get_cmap('tab20', len(temperature_thresholds))

for threshold in temperature_thresholds:
    # Calculate the number of days above the threshold for each year
    yearly_counts = df4[df4['Heat_Index'] > threshold].groupby('Year')['Date'].count()
    ax.bar(yearly_counts.index, yearly_counts, label=f'{threshold}°F', color=colors(temperature_thresholds.index(threshold)))

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Number of Days above HI Thresholds')
plt.title('Number of Excessive Heat Days at KMIA during Summer 2019-2023')

# Show the legend
plt.legend(title='Heat Index (HI)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
