```python
import pandas as pd
import datetime as dt
import plotly.express as px
```

```python
!pip install plotly
```

```python
df = pd.read_csv('Heat-Stats_KMIA.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year.astype(str)
df['mthday']=df['Date'].dt.strftime('%d-%b')
df['mthday']=pd.to_datetime(df['mthday'],format='%d-%b',errors='coerce')
df=df.sort_values(by='mthday')
df['avg_daily'] = df.groupby('mthday')['Air_Temp'].transform('mean')

plot=px.scatter(df,x='mthday',y='Air_Temp',color='Year',
             title = 'Maximum Observed Air Temperature during Summer at KMIA (2019-2023)',
                color_discrete_sequence=px.colors.qualitative.Set2, template='simple_white',
                hover_data={'mthday':False})
plot.add_scatter(x=df['mthday'], y=df['avg_daily'], mode='lines', line_color='darkblue', name='Mean')
plot.update_layout(xaxis_tickformat='%d-%b', xaxis_title='Date', yaxis_title='Air Temperature (degF)',
                   title_x=0.5, hovermode='x unified')
plot.update_traces(hovertemplate='%{y}')
plot.show()
```

```python
py.plot(plot, filename='Maximum Air Temperature KMIA 2019-2013', auto_open=False)
```

```python
df2 = pd.read_csv('Heat-Stats_KMIA.csv')
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
df2['Year'] = df2['Date'].dt.year.astype(str)
df2['mthday']=df2['Date'].dt.strftime('%d-%b')
df2['mthday']=pd.to_datetime(df2['mthday'],format='%d-%b',errors='coerce')
df2=df2.sort_values(by='mthday')
df2['avg_daily'] = df2.groupby('mthday')['Heat_Index'].transform('mean')

plot2=px.scatter(df2,x='mthday',y='Heat_Index',color='Year',
             title = 'Maximum Observed Heat Index during Summer at KMIA (2019-2023)',
                color_discrete_sequence=px.colors.qualitative.Pastel1, template='simple_white',
                hover_data={'mthday':False})
plot2.add_scatter(x=df2['mthday'], y=df2['avg_daily'], mode='lines', line_color='darkblue', name='Mean')
plot2.update_layout(xaxis_tickformat='%d-%b', xaxis_title='Date', yaxis_title='Heat Index (degF)',
                   title_x=0.5, hovermode='x unified')
plot2.update_traces(hovertemplate='%{y}')
plot2.show()
```
